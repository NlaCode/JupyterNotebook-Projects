```python
plt.figure(figsize=(10,10))
plt.title('Correlation Map')
ax=sns.heatmap(spoti.corr(),
               linewidth=3.1,
               annot=True,
               center=1)
```


Também conhecido como heatmap, demonstra a relevância da váriavel de acordo com a intensidade da cor.
Quanto mais forte for, maior é sua relevância. 
No qual Loudness db e Energy apresentam o maior nível de correlação no mapa.

```python
sns.catplot(y = "Genre", kind = "count",
            palette = "pastel", edgecolor = ".6",
            data = spoti)
plt.show()
```

Dance pop e pop estão com maior representatividade no top spotify list,
demonstrando a tendência daquele ano.

```python
plt.figure(figsize=(12,12))
sns.jointplot(x=spoti["Beats.Per.Minute"].values, y=spoti['Popularity'].values, height=10, kind="kde",)
plt.ylabel('Popularity', fontsize=12)
plt.xlabel("Beats.Per.Minute", fontsize=12)
plt.title("Beats.Per.Minute Vs Popularity", fontsize=15)
plt.show()

```

Este gráfico está demonstrando a conexão entre as batidas por minuto e a popularidade da música. Concluindo que não necessáriamente quanto mais batidas a música tem, maior será sua popularidade. Realçando o gráfico anterior, pois os gêneros musicais tecno e house, não possuem tanta popularidade .

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Loudness..dB..', y='Popularity', data=spoti)
plt.xlabel('Loudness..dB..', fontsize=12)
plt.ylabel('Popularity', fontsize=12)
plt.show()

```

Este gráfico demonstra a relação entre popularidade e o volume em decibéis. O qual podemos concluir que não necessáriamente a música mais alta é a mais popular.
